```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Set up the driver (you can add your path to ChromeDriver if needed)
options = Options()
options.add_argument('--headless')  # run in headless mode
options.add_argument('--disable-gpu')
driver = webdriver.Chrome(options=options)  # Make sure ChromeDriver is installed and in PATH
```

```python
def get_article_links(base_url):
    driver.get(base_url)
    time.sleep(5)  

    for _ in range(2):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)

    links = set()
    article_elements = driver.find_elements(By.CSS_SELECTOR, 'a.title-link')  
    for elem in article_elements:
        url = elem.get_attribute('href')
        if url and '/bangladesh/crime/' in url:
            if url.startswith('/'):
                url = 'https://www.prothomalo.com' + url
            links.add(url)
    return list(links)
```

```python
base_url = 'https://www.prothomalo.com/bangladesh/crime'
article_links = get_article_links(base_url)
print(article_links)
```

```python
def extract_sentences_from_article(url):
    driver.get(url)
    time.sleep(3)

    paragraphs = driver.find_elements(By.CSS_SELECTOR, 'div.story-element.story-element-text p')

    full_text = ' '.join(p.text for p in paragraphs if p.text.strip())

    sentences = [s.strip() + '।' for s in full_text.split('।') if s.strip()]
    return sentences
```
